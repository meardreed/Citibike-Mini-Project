```python
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
file = 'citibike-nov22.csv'
# The correct encoding must be used to read the CSV in pandas
nov_df = pd.read_csv(file, encoding="ISO-8859-1")
nov_df.head()
```

```python
# Name of the CSV file
file2 = 'citibike-dec22.csv'
# The correct encoding must be used to read the CSV in pandas
dec_df = pd.read_csv(file2, encoding="ISO-8859-1")
dec_df.head()
```

```python
nov_df.count()
```

```python
dec_df.count()
```

```python
# Concatenate all dataframes into one
combined_df = pd.concat([nov_df, dec_df], axis=0)
combined_df
```

```python
# Save the dataframe to a CSV file
combined_df.to_csv('citibike.csv', index=False)
```
